```python
 #model.py - DATA MODEL
class SolarData:
    def _init_(self, timestamp, solar_power):
        self.timestamp = timestamp
        self.solar_power = solar_power

class WindData:
    def _init_(self, timestamp, wind_power):
        self.timestamp = timestamp
        self.wind_power = wind_power
```

```python
# planner_logic.py — ENERGY PLANNER
def best_solar_hours(df, threshold=500):
    return df[df['solar_power'] > threshold]
def best_wind_hours(df, min_speed=5):
    return df[df['wind_speed'] > min_speed]

```

```python
#shopping_list.py — SUMMARIZE DATA
def total_energy(df):
    return df['solar_power'].sum(), df['wind_power'].sum()
```
